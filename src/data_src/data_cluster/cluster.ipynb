```python
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import math
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 


```

```python
data_path = '../../../data/'
output_path = data_path + 'result/draw/'

total = pd.read_excel(data_path + '10_day.xlsx')
print(total.info)
```

```python
str1 = 'weekday'
str2 = 'weekend'

strArr = []
size = 2

strArr.append(str1)
strArr.append(str2)

pstr = 'per_'
#pstr = ''

```

```python
kmeans = KMeans(n_clusters=5)

points = pd.DataFrame(total[pstr+strArr[0]]).reset_index(drop=True)

for idx in range(1,size):
    points[strArr[idx]] = total[pstr+strArr[idx]]

points.columns = strArr


kmeans.fit(points)

result_by_sklearn = points.copy()
result_by_sklearn["cluster"] = kmeans.labels_
result_by_sklearn.head(n=10)
```

```python
sns.pairplot(result_by_sklearn,hue='cluster',palette='Set2')
plt.show()
```

```python
result_by_sklearn.to_excel(output_path + 'result.xlsx')
```
